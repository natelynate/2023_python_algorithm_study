```python
"""
Q15 특정 거리의 도시 찾기(1) https://www.acmicpc.net/problem/18352

솔루션: 위치X에서 거리K인 노드들을 반환하는 문제이므로 연결된 그래프들을 탐색해야 함.

x에서부터 최단거리가 m인 노드들을 찾는 문제는 DFS랑은 안 어울리나? 노드k로 접근하는 경로가 1개 이상이라고 할 때 제일 탐색되는 경로가 최단 경로라는 
보장이 없기 때문... 그렇다고 경로를 1개 이상 카운트하고 min()을 취하는 순간 전체 그래프에서 순환되는 지점이 하나라도 있으면 무한재귀에 늪에 빠진다.  
그래서 DFS를 사용할 수는 없고 BFS를 써야 할 듯?

BFS를 사용한다면 연결 정보들을 따로 리스트로 저장해놓고 출발점 X를 queue에 넣고 BFS탐색을 시작. 
queue에 넣을 때는 (도착노드, X로부터의 거리)의 튜플 형식으로
넣고 거리가 == K인 조건을 매 인스턴스의 초반에 넣어서 조건이 참이면 종료하고 결과를 따로 리스트에 저장한 후 리스트 정렬 후 반환.
"""
from collections import deque
import sys

N, M, K, X = map(int, sys.stdin.readline().split())      # sys.stdin.readline().split()
visited = [False for i in range(N+1)]
edges = [[] for _ in range(N+1)]


# 간선 정보 입력
for _ in range(M):
    edge = list(map(int, sys.stdin.readline().split()))
    edges[edge[0]].append(edge[1]) # 각 노드만큼 내부 리스트를 만들고 연결된 노드들을 등재함  

distance = [-1] * (N + 1) # 도시 거리 리스트
distance[X] = 0 # 자기자신과의 거리는 0

queue = deque([X])   

while queue:
    # 매 인스턴스의 노드에 연결되어 있는 노드들 중 미방문인 애들을 queue에 append하고 distance를 갱신해준다
    start_node = queue.popleft()
    for next_node in edges[start_node]:
        if distance[next_node] == -1:
            queue.append(next_node)
            distance[next_node] = distance[start_node] + 1

# 전체 노드의 거리 리스트에서 거리가 K인 노드의 번호들만 남김
k_list = [index for index, value in enumerate(distance) if value == K]

# 만약 K_list에 원소가 없으면 -1 반환
if not k_list:
    print(-1)

# 원소가 1개라도 있으면 모두 차례대로 출력(인덱스=노드번호이므로 이미 상향정렬됨)
else:
    for node in k_list:
        print(node)
```

```python
"""
Q16 연구소 https://www.acmicpc.net/problem/14502

BFS 솔루션:
바이러스 확산 과정을 BFS를 통해 모델링하는 시뮬레이션 구현 문제이다. 빈 칸들 중 임의의 가벽 3개를 세울 수 있는 모든 가짓수를 직접 시뮬레이션하면서 
빈 칸 개수를 세서 만약 저장해놨던 최소값보다 더 적은 개수가 나오면 그것으로 대치하는 방식으로 구현하면 될 것이다.
추후에 combinations 함수를 활용할 때 편의성을 위해 지도의 행을 입력받을 때 for문으로 빈 공간(0)과 바이러스(2)의 위치를 별개의 리스트에 저장하는 방식을 취함.

원래는 queue를 매번 while문의 인스턴스마다 생성했는데 queue는 loop 바깥에서 생성하고 인스턴스 시작시 
새로운 바이러스 위치들을 넣고 BFS로 탐색하면 똑같은 결과를 효율적으로 낼 수 있음.
원래는 python3 기준 3776ms고 queue 생성을 밖으로 옮긴 버전에서는 3660ms 인데 아주 큰 차이는 아니긴 함.
"""
from itertools import combinations
from collections import deque

N, M = map(int, input().split())
virus_cells= list()
empty_cells = list() 
graph = list() 
max_safe_zone = -1 # 안전지대 크기()
dy = [-1, 1, 0, 0] # 상하좌우
dx = [0, 0, -1, 1] 

# 데이터 입력
for y in range(N):
    row = list(map(int, input().split()))
    for x, cell in enumerate(row):
        if cell == 2: # 
            virus_cells.append((y, x))
        elif cell == 0:
            empty_cells.append((y, x))
    graph.append(row)

queue = deque() # BFS를 위한 queue 생성

# 가벽 세 개를 세우는 경우의 수를 itertools의 combinations로 계산 -> 가벽은 빈 칸에만 세울 수 있으므로 빈 칸 리스트에서 3개를 뽑는 경우의 수를 계산한다.
for wall_loc in combinations(empty_cells, 3):
    safe_zone = len(empty_cells) - 3 # 바이러스가 퍼지기 전의 안전지대 크기 

    # 시뮬레이션 전 가벽 3개를 세움
    for wall in wall_loc: 
        wall_y, wall_x = wall
        graph[wall_y][wall_x] = 1 
    
    # 현재 바이러스의 위치를 queue에 삽입
    for virus in virus_cells:
        queue.append(virus)

    # queue에 소진될 때까지 BFS 시작
    while queue:
        y, x = queue.popleft()
        for dir in range(4): # 임의의 위치에서 4방향으로 이동이 가능한지 확인
            ny = y + dy[dir]
            nx = x + dx[dir]
            if (0 <= ny < N) and (0 <= nx < M) and (graph[ny][nx] == 0): # 새로운 칸이 지도 내부의 빈 칸이라면 바이러스가 확산됨
                graph[ny][nx] = 2
                queue.append((ny, nx))
                safe_zone -= 1 # 바이러스가 퍼질 때마다 시작 당시의 안전지대 개수에서 1을 차감
    max_safe_zone = max(safe_zone, max_safe_zone) # 새로운 빈 칸 개수와 기존의 빈 칸 개수 중 max를 취함

    # 해당 인스턴스에서 설치한 가벽 제거
    for wall in wall_loc: 
        graph[wall[0]][wall[1]] = 0
    
    # 바이러스 제거
    for empty_tile in empty_cells:
        graph[empty_tile[0]][empty_tile[1]] = 0

print(max_safe_zone)
```

```python
"""
Q16 연구소 https://www.acmicpc.net/problem/14502

DFS 솔루션:

위의 솔루션에서 탐색 부분만 DFS식 재귀함수로 구현한 버전. DFS를 사용했을 때 구현 실력 부족으로 BFS만큼 세련되게 집계 등을 못하고 추가적인 절차가 들어가긴 했는데
그럼에도 불구하고 파이썬 기준 2052ms로 꽤나 유의미한 시간 감소 효과를 봤다 (?!)
"""
from itertools import combinations

def dfs(graph, y, x):
    if (N <= y) or (0 > x) or (0 > y) or (M <= x) or (graph[y][x] != 0): # dfs 종료조건
        return
    graph[y][x] = 2
    for dir in range(4):
        ny = y + dy[dir]
        nx = x + dx[dir]
        dfs(graph, ny, nx)

N, M = map(int, input().split())
virus_cells= list()
empty_cells = list() 
graph = list() 
max_safe_zone = -1 # 안전지대 크기()
dy = [-1, 1, 0, 0] # 상하좌우
dx = [0, 0, -1, 1] 

# 데이터 입력
for y in range(N):
    row = list(map(int, input().split()))
    for x, cell in enumerate(row):
        if cell == 2: # 
            virus_cells.append((y, x))
        elif cell == 0:
            empty_cells.append((y, x))
    graph.append([i if i != 2 else 0 for i in row])

for wall_loc in combinations(empty_cells, 3):
    for wall in wall_loc: 
        wall_y, wall_x = wall
        graph[wall_y][wall_x] = 1 

    # dfs탐색 시작
    for virus in virus_cells:
        dfs(graph, virus[0], virus[1])
    
    # 빈 칸 개수 집계
    empty_count = sum([sum([1 for cell in row if cell == 0]) for row in graph])
    max_safe_zone = max(max_safe_zone, empty_count)

    # 지도 원복
    for wall in wall_loc: 
        graph[wall[0]][wall[1]] = 0
    for tile in empty_cells:
        graph[tile[0]][tile[1]] = 0
    for virus in virus_cells:
        graph[virus[0]][virus[1]] = 0
    
else:
    print(max_safe_zone)
```

```python
"""
Q17 경쟁적 전염
https://www.acmicpc.net/problem/18405

소요시간: 15:35~16:22 
메모리(KB): 34184
시간(MS): 164

솔루션 1:
간단한 BFS 시뮬레이션으로 풀 수 있다. 규칙대로 그대로 구현한 후 graph에서 특정 위치의 값을 반환하면 정답임. queue에 아예 현재 시간을 포함시킨 튜플을 넣은 후 매 인스턴스마다 종료조건으로
시간이 == S인지 확인해서 만족하면 탐색을 종료하는 방식으로 구현했다.
"""
from collections import deque

N, K = map(int, input().split())
petri_tube = [list(map(int, input().split())) for _ in range(N)] 
S, X, Y = map(int, input().split())
prequeue = list()
dx = [0, 0, -1, 1] # 상하좌우
dy = [-1, 1, 0, 0]

# 전체 공간을 iterate하면서 바이러스가 있는 타일들을 다 queue에 삽입
time = 0
for i in range(N):
    for j in range(N):
        if petri_tube[i][j] != 0: # 빈 칸이면 아니면 바이러스이므로 (바이러스 번호, y좌표, x좌표, 시간) 튜플 형식으로 queue에 넣는다. 
            prequeue.append((petri_tube[i][j], i, j, time))

prequeue = sorted(prequeue) # 번호가 낮은 것부터 증식하니까 prequeue를 상향으로 정렬
queue = deque(prequeue)
#print(queue)

while queue:
    virus, y, x, time = queue.popleft()
    if time == S:
        break
    
    # 4방향으로 증식
    for dir in range(4):
        ny = y + dy[dir]
        nx = x + dx[dir]
        if (ny >= N) or (ny < 0) or (nx < 0) or (nx >= N): # 시험관 범위를 벗어날 경우
            continue
        
        if petri_tube[ny][nx] != 0: # 새로운 타일에 이미 다른 바이러스가 있거나 
            continue

        petri_tube[ny][nx] = virus # 맵에 바이러스 타일 추가
        queue.append((virus, ny, nx, time+1)) # 새로운 타일을 queue에 추가

print(petri_tube[X-1][Y-1])
```

```python
"""
Q17 경쟁적 전염
https://www.acmicpc.net/problem/18405

소요시간: 
메모리(KB): 31256
시간(MS): 64

솔루션 2:
바이러스가 순서대로 동시에 상하좌우 방향으로 확산하고 서로 덮어씌워지지 않으므로, 
논리적으로 X, Y 타일에 가장 먼저 도착하게 될 바이러스는 X, Y타일과의 거리가 가장 가까우면서 번호가 낮은
바이러스라고 추론할 수 있다. 그래서 각 바이러스의 좌표와 (X, Y) 간의 맨하탄 거리를 기준으로 바이러스를 정렬하고, 
만약 맨하탄거리가 같은 바이러스가 2개 이상이면 그 중에서 낮은 번호를 
반환하도록 하면 시뮬레이션할 필요 없이 analytical하게 답을 구할 수 있다.

이 경우 queue를 가지고 직접 bfs를 했을 때 164ms, 맨하탄거리 기준으로 정렬하는 방법을 사용했을 땐 68ms로 확연히 빨라짐.
"""

def get_manhattan(a: tuple, b:tuple) -> tuple:
    return (abs(a[0] - b[0]) + abs(a[1] - b[1]))


N, K = map(int, input().split())
petri_tube = [list(map(int, input().split())) for _ in range(N)]
S, X, Y = map(int, input().split())

distances = list() # 바이러스들의 맨하탄 거리를 저장할 리스트
viruses = list() # 바이러스 좌표 저장용 리스트

# 전체 맵을 순회하면서 바이러스의 맨하탄 거리를 각각 계산해서 distances에 삽입
for i in range(N):
    for j in range(N):
        if petri_tube[i][j] != 0:
            viruses.append((petri_tube[i][j], get_manhattan((X, Y), (i+1, j+1)))) # (바이러스 번호, (X,Y)와의 맨하탄 거리)

viruses = sorted(viruses, key=lambda x: (x[1], x[0])) # 인덱스 1원소 기준으로 정렬한다. 


# 가장 낮은 번호이면서 목표지점과의 맨하탄거리가 짧은 바이러스를 반환하고, 맨하탄거리 = X,Y에 도달하는 시간 과 같으므로 만약 맨하탄거리가 S 이상이면 S초 이내에 도달 못한단 뜻이므로 0을 대신 반환
print(viruses[0][0] if viruses[0][1] <= S else 0) 
```

```python
"""
Q19 연산자 끼워넣기 # 개선필요

소요시간: 약 1시간
메모리(KB): 218344
시간(MS): 	4124

DFS와 BFS를 모두 사용할 수 있을 것 같음. 연산자를 매번 하나씩 순서대로 사용하면서 사용할 때마다 사용 가능한 연산자를 하나 제외해가면서 시뮬레이션을 하되 
매번 연산자를 모두 사용했는지 여부를 종료조건으로 확인하면서 종료되면 global variable result에 결과값을 저장하고 모든 탐색이 끝났을 때 max(), min()을 써서 결과를 반환함. 
즉 DFS에서는 현재 인스턴스까지 사용한 연산자 리스트, 그리고 이번에 사용할 수 있는 연산자들을 리스트로 담고 돌아가면서 매번 사용할 수 있는 연산자들의 최대 경우의 수 만큼의 넓이로 분기된다. 

많이 느린 편. pypy3에서도 4124ms가 나오고 python3에서는 통과못함.
sys.stdin.readline()을 쓰면 3692ms로 소폭 줄어들긴 함..
"""
N = int(input())
num = list(map(int, input().split()))
operators = list(map(int, input().split()))
used_operators = []
unused_operators = []
for idx, val in enumerate(operators):
    if val != 0:
        for _ in range(val):
            unused_operators.append(idx)

results = set()

def dfs(used_operators, unused_operators):
    if len(used_operators) == N-1:
        result = 0
        result += num[0]
        for i in range(len(used_operators)):
            if used_operators[i] == 0:
                result += num[i+1]
            elif used_operators[i] == 1:
                result -= num[i+1]
            elif used_operators[i] == 2:
                result *= num[i+1]
            elif used_operators[i] == 3:
                result = int(result / num[i+1])
        results.add(result)
        return
    
    for operator in unused_operators:
        n_used_operators = [i for i in used_operators]
        n_unused_operators = [i for i in unused_operators]

        n_used_operators.append(operator)
        n_unused_operators.remove(operator)

        dfs(n_used_operators, n_unused_operators)


dfs(used_operators, unused_operators)
print(max(results), min(results), sep='\n')
```

```python
"""
Q19 연산자 끼워넣기(답지)

소요시간: 
메모리(KB): 
시간(MS):

DFS에서 매 인스턴스에 "사용한 연산자", 사용안한 연산자 2개를 pass down하는 번거로운 방식 말고
1. 현재까지의 계산결과 (int)
2. 사용한 연산자 개수 (int)
를 passdown하면서 하면 매번 dfs를 traverse할 때마다 새로운 리스트를 선언할 필요가 없기 때문에 훨씬 시간을 줄일 수 있다. 
사용할 수 있는 연산자들은 global variable로 빼버린 다음에 관리하면 됨. 

"""
N = int(input())
num = list(map(int, input().split()))
add, sub, mul, div = map(int, input().split()) # 남은 연산자 개수를 표시하는 정수값을 담은 변수들을 선언
min_value = 1e99
max_value = -1e99

def dfs(level:int, result:int):
    global add, sub, mul, div, min_value, max_value
    if level == N: # 만약 최대 연산자 개수만큼 반복했다면 종료
        max_value = max(result, max_value)
        min_value = min(result, min_value)
        
    else: # DFS 탐색
        if add > 0: # 만약 사용하지 않은 + 연산자가 1개라도 있다면:
            add -= 1 # add의 수를 줄여줘야 바로 직후의 멀티버스를 돌릴 때 + 연산자 개수가 정상적임
            dfs(level+1, result+num[level])
            add += 1 # 바로 위의 멀티버스가 정지한 직후에는 첫 +를 사용하지 않은 멀티버스여야 하므로 다시 개수를 +1해서 원복시킴

        if sub > 0:
            sub -= 1
            dfs(level+1, result-num[level])
            sub += 1

        if mul > 0:
            mul -= 1
            dfs(level+1, result*num[level])
            mul += 1

        if div > 0:
            div += 1
            dfs(level+1, result // num[level])
            div += 1

dfs(1, num[0])
print(max_value, min_value, sep='\n')
```

```python
"""
Q19 연산자 끼워넣기

소요시간: 약 1시간
메모리(KB): 115268
시간(MS): 1112

오히려 그냥 permutations를 써서 완전탐색을 하는 것이 4배 가량 빠르다. 코드도 짧고...
"""
from itertools import permutations

N = int(input())
integers = list(map(int, input().split()))
operator_counts = list(map(int, input().split()))  # +, -, *, /
operator_list = ['+', '-', 'x', '/']
operator_seq = []

for i in range(len(operator_counts)):
    for j in range(operator_counts[i]):
        operator_seq.append(operator_list[i]) 

max_ = -9999999999
min_ = 999999999999

for com in permutations(operator_seq, N-1): # 각 연산자 조합마다 실시
    result = integers[0] # 0 번째는 일단 갖고 시작
    for i in range(1, N):
        if com[i-1] == '+':
            result += integers[i]
        elif com[i-1] == '-':
            result -= integers[i]
        elif com[i-1] == 'x':
            result *= integers[i]
        elif com[i-1] == '/':
            result = int(result / integers[i])

    if result > max_:
        max_ = result
    if result < min_:
        min_ = result
        
print(max_)
print(min_)
```

```python
"""
Q20 감시피하기
https://www.acmicpc.net/problem/18428

소요시간: 16:30~
메모리(KB): 35560
시간(MS): 128

솔루션:
이것도 경쟁적전염과 똑같이 combinations로 3개의 장애물에 배치해놓고 bfs로 확인하면서 하나라도 적발되는 학생이 있으면 조기종료하는 식으로 구성할 수 있다. 
다만 단, 전체 선생님의 수는 5이하의 자연수, 전체 학생의 수는 30이하의 자연수이므로 '학생'들을 가지고 선생님한테 '걸리는지' 보는 것보다 '선생님'을 기준으로 1명이라도 '잡는지' 검사하는 것이
조금 더 빠르리라고 추측함. 

독립적인 인스턴스를 위해 원본 맵의 deepcopy 사본을 매 인스턴스마다 생성하고 사용하는 것보다 원본에다가 수정을 하고 인스턴스 종료 전에 원복시켜놓는 방식이
더 효율적이라는 걸 이전 문제에 확인했으므로 그 부분을 반영함. + queue 오브젝트도 매번 새로 생성하는 것이 아니라 재활용할 수 있게 함.

전에는 156,152ms 걸렸는데 128ms로 줄였으니까 발전...? 이려나
"""

from itertools import combinations # 가벽을 세우는 3개의 위치는 순서가 달라도 동일한 의미이므로 combinations 사용함
from collections import deque

N = int(input())
teachers = list()
empty_tiles = list()
worldmap = []
result = 0

dir_x = [0, 0, -1, 1] # 상하좌우
dir_y = [-1, 1, 0, 0] 

for idx_i in range(N):
    row = input().split()
    for idx_j, j in enumerate(row):
        if j == 'T':
            teachers.append((idx_i, idx_j))
        elif j == 'X':
            empty_tiles.append((idx_i, idx_j))
    worldmap.append(row)

queue = deque()
caught = False
for walls in combinations(empty_tiles, 3):
    caught = False
    # 아직까지 잡힌 학생이 없을 경우 탐색 시작  
    for wall in walls:
        worldmap[wall[0]][wall[1]] = 'W' # 3곳의 임의의 위치에 가벽 생성

    for teacher in teachers:
        for dir in range(4):
            queue.append((teacher[0], teacher[1], dir)) # y 좌표, x 좌표, 방향integer를 튜플로 queue에 삽입
            
    while queue: # BFS시작
        y, x, dir = queue.popleft()
        ny = y + dir_y[dir]
        nx = x + dir_x[dir]
        if not (0 <= ny < N) or not (0 <= nx < N):
            continue

        if (worldmap[ny][nx] == 'W'): # 벽이나 다른 선생님 타일일 경우
            continue

        elif worldmap[ny][nx] == 'S': # 학생이면 queue 종료
            caught = True # 잡힘 여부 갱신
            # print("잡힘", f"{ny}, {nx}")
            break
        
        # 빈 타일일 경우 타일 종류를 T로 바꾸고 다음 타일을 queue에 추가
        worldmap[ny][nx] = 'T'
        queue.appendleft((ny, nx, dir))

    else: # while-else문을 써서 만약 모든 while문이 중간에 조기종료되지 않고 돌았으면 -> 하나도 안 걸렸단 뜻. 즉 caught=False가 유지됨. 
        result = 1
        break

    # 다음 인스턴스 시작 전에 worldmap을 원상복구해놓음 
    for wall in walls:
        worldmap[wall[0]][wall[1]] = 'X' # 가벽 해체
    
    for cell in empty_tiles:
        worldmap[cell[0]][cell[1]] = 'X' # 빈 칸 복원

if caught:
    print("NO")
else: 
    print("YES")
```

```python
""""
Q21. 인구이동 https://www.acmicpc.net/problem/16234
소요시간:
메모리(KB): 
시간(MS): 

핵심은 NxN을 전부 BFS로 살펴보면서 연합이 아예 없는 상태가 될 때까지 전체 순환을 반복하는 것.
즉 가장 상위에서는 while True문으로 반복하고, 매 인스턴스마다 전체 맵을 BFS로 순환하면서 연합이 형성되는지 BFS로 확인한다.
BFS에서는 결성된 연합을 리스트로 보관하고, 이후 인구 분배를 진행한 후 다시 상위의 while True문으로 복귀. 만약 BFS로 전체를 확인했는데
결성 가능한 연합의 수가 0일 경우에는 종료하고 일수를 반환하면 될 것이다.   

N = 정사각형 땅의 한 변 길이
L, R = 인구이동 발생 조건의 최소값, 최대값

기존에 있었던 오류는 아마 for반복문 내에서 for i in iterable할 때 i를 뭐라고 하는 지 모르겠네..하여튼 받아오는 변수와 같은 글자를 
for 문 내부의 다른 for문 내에서 사용하면서 동일한 symbol을 사용했기 때문에 값이 덮어씌워져 버렸고 그래서 다른 값들이 뒤섞여버려서 발생한 문제였다.
거진... 10시간 동안의 디버깅 끝에야 발견한 사실...당연히 동일한 글로벌 네임스페이스를 공유한다...
for문의 variable과 같은 것을 사용하지 않도록 앞으로도 조심 또 조심...
리스트 컴프리헨션 표현식 내부에 있으면 별도의 네임스페이스를 지니는 것으로 보임 << 나중에 더 이론 알아보고
리스트 컴프리헨션에서 같은 네임을 사용해도 for문은 정상적인 순서의 값들을 여전히 yield한다. 즉 문제 코드에서 문제가 발생한 지점은 
for y, x in country: 부분이었다..여기서 country내부에 있는 튜플내의 원소값들이 y,x에 덮어씌워진 것..

발?전한 점:
   처음에는 정말 속상하고 화가 머리끝까지 났지만 8시간 쯤이 지난 뒤에는 더 이상 화가 나는 게 아니라 이 기이함의 원천이 무얼까 오히려 궁금해졌다... 
   그리고 얼마 안 가서 문제점을 찾은 걸 보아...(그냥 변수명 문제인가 싶어서 변수명들을 그냥 임의로 순서 등등을 이리저리 바꿔보다가 어쩌다 보니 원래 제출하면 맞는 답변이
   오류가 되는 걸 발견했었고 거기서부터 추적한 끝에 알아냈음)...화를 내지말고 다음부터는 궁금해하는 걸 먼저 하기로... 명심하자...상대는 기계다..
"""
from collections import deque

N, L, R = map(int, input().split())
worldmap = [list(map(int, input().split())) for _ in range(N)]

dy = [-1, 1, 0, 0]
dx = [0, 0, -1, 1]

def bfs(y, x):
    queue = deque()
    union = []
    union.append((y, x))
    queue.append((y, x))

    while queue:
        y, x = queue.popleft()
        for i in range(4): # 임의의 타일에서 상하좌우 4방향으로 인접타일을 탐색
            ny = y + dy[i]
            nx = x + dx[i]
            if 0 <= ny < N and 0 <= nx < N and visited[ny][nx] == 0:
                if L <= abs(worldmap[ny][nx] - worldmap[y][x]) <= R:
                    visited[ny][nx] = 1
                    queue.append((ny, nx))
                    union.append((ny, nx))                 
    return union

days = 0 
while True:
    visited = [[0 for _ in range(N)] for _ in range(N)]
    flag = 0
    for i in range(N):
        for j in range(N):
            if visited[i][j] == 0:
                visited[i][j] = 1
                country = bfs(i, j)

                if len(country) > 1:
                    flag = 1
                    avg_pop = sum([worldmap[y][x] for y, x in country]) // len(country)
                    
                    for y, x in country:
                        worldmap[y][x] = avg_pop
                        
    if flag == 0:
        print(days)
        break
    days += 1
```

```python
"""
Q22 블록이동하기
https://school.programmers.co.kr/learn/courses/30/lessons/60063

소요시간: 
메모리(KB): 
시간(MS): 

로봇이 움직일 수 있는 모든 방식을 BFS의 매 인스턴스때마다 모두 수행하고, 종료조건을 매번 확인해서(종교조건=두 cell의 어떤 좌표들 중 하나라도 목표 좌표와 일치)
일치하는 순간 queue의 원소에 포함된 time의 값을 확인해서 그때마다 최소값으로 갱신해주는 것을 계속 반복한다. 

로봇이 두 칸을 차지하므로 기존 방식처럼 boolean 값을 가진 map과 동일한 크기의 visited list를 만들어서 관리하는 것이 힘듦 
-> 그래서 효율성을 희생하는 대신 로봇이 있었던 좌표 2곳을 튜플로 한 번에 묶어서 모두 다 저장하도록 한 뒤 [(a, b), [c, d]] in list()를 통해 확인하는 
방법을 취했음. 

원래는 로봇이 현재 가로방향인지 세로방향인지 파악하는 것을 instance를 pop하고 나서 y좌표 기준으로 후시적으로 확인했는데,
아예 queue에 넣는 원소에 방향을 표현하는 코드를 포함시키면 작업 가짓수를 줄일 수 있을 것. 
여기서 방향 0은 가로. 1은 세로를 나타낸다. 

로봇의 이동은 직선이동 4가지, 제자리회전 4가지를 구분해서 동작한다.

테스트 13 > 실패 (시간 초과)
"""
from collections import deque

def solution(board):
    N = len(board)
    # 방향벡터
    dy = [-1, 1, 0, 0]
    dx = [0, 0, -1, 1]

    # 로봇의 초기 위치와 시작시간 time=0, 방향 0을 넣은 튜플을 queue에 최초 원소로 넣고 생성
    queue = deque([[0, 0, 0, 1, 0, 0]])
    
    # visited 리스트에 로봇이 초기 위치 튜플을 삽입
    visited = [] 
    visited.append((0, 0, 0, 1))
    
    # for i in board:
    #    print(i)
    while queue:
        y1, x1, y2, x2, time, align = queue.popleft()
        print(y1, x1, y2, x2, time)
        # 종료조건
        if (y1 == N-1 and x1 == N-1) or (y2 == N-1 and x2 == N-1): # 로봇의 몸체 중 한 블록이라도 종착점에 도착했는지 확인한다.
            return time
        
        # 방향에 따라 탐색 시작
        if align == 1:
            pass

        elif align == 0:
            pass




        # 세로방향일 경우
        if align == 1:
            # 상하 이동
            for i in [-1, 1]:
                if (0 <= y1+i < N) and (board[y1+i][x1] == 0) and ((y1-i, x1), (y2+i, x2)) not in visited:
                    queue.append((y1+i, x1, y1, x1, time+1, align))
                    visited.append((y1+i, x1, y1+i, x1))
            # 좌우 이동
            # 세로로 전개된 로봇의 양 옆 두 칸이 모두 비어 있다면 수평이동을 포함해서 회전이동도 할 수 있음. 즉 queue에 3개의 원소가 추가된다.
            for i in [-1, 1]:
                if (0 <= x1+i < N) and (0 <= y1+i < N) and (0 <= x2+i < N) and (0 <= y2+i < N):
                    queue.append((y1, x1+i, y2, x2+i, time+1, 1 if align == 1 else 0)) # 직선 이동
                    queue.append((y1, x1, y1+i, x1+i, time+1, 1 if align == 1 else 0)) 
                    queue.append((y2, x2, y2+i, x2+i, time+1, 1 if align == 1 else 0)) 

                    visited.append(((y1, x1+i), (y2, x2+i)))
                    visited.append(((y1, x1), (y1+i, y1+i)))
                    visited.append(((y2, x2), (y2+i, x2+i)))

        # 가로 방향일 경우
        elif align == 0: # 가로방향일 경우:
            # 상하 이동
            # 가로로 전개된 로봇의 위 아래 두 칸이 모두 비어있으면 수평이동에 더해 회전이동이 가능함
            for i in [-1, 1]:
                if (0 <= x1+i < N) and (0 <= y1+i < N) and (0 <= x2+i < N) and (0 <= y2+i < N) and (board[y1+i][x2] == 0) and (board[y2+i][x2] == 0) and (((y1+i, x1), (y2+i, x2)) not in visited):
                    queue.append((y1+i, x1, y2+i, x2, time+1, 1 if align == 1 else 0)) # 직선 이동
                    queue.append((y1, x1, y1+i, x1+i, time+1, 1 if align == 1 else 0)) 
                    queue.append((y2, x2, y2+i, x2+i, time+1, 1 if align == 1 else 0)) 

                    visited.append(((y1+i, x1), (y2+i, x2)))
                    visited.append(((y1, x1), (y1+i, x1+i)))
                    visited.append(((y2, x2), (y2+i, x2+i)))
        

solution([[0, 0, 0, 1, 1], [0, 0, 0, 1, 0], [0, 1, 0, 1, 1], [1, 1, 0, 0, 1], [0, 0, 0, 0, 0]])
```
