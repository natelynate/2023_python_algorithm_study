```python
"""
Q27 정렬된 배열에서 특정 수의 개수 구하기 

소요시간: 
메모리:
시간:

솔루션:
이미 상향정렬되어 있으므로 이진탐색을 사용할 수 있음. 값이 x인 원소 n개를 찾아야 하므로, 찾아야 하는 것은 가장 왼편에 있는 x의 인덱스, 그리고 가장 우측에
있는 x의 인덱스를 찾을 수 있으면 서로 인덱스를 빼서 x가 몇 개 있는지 알 수 있다. 

양극단에 위치했는지 여부는 각각 좌측과 우측의 원소 값을 확인해서 어느 한 쪽이 자기자신이 아니라면 = 극단에 위치해 있다고 볼 수 있다. 
"""

#N, x = map(int, input().split())
#seq = list(map(int, input().split()))
N, x = 7, 2
seq = [1, 1, 2, 2, 2, 2, 3]

def find_end(start:int, end:int, target:int) -> int:
    while start < end:
        mid = (end + start) // 2 # 중간점 인덱스 계산
        if seq[mid] > target: # 만약 중간점의 값이 target보다 크다면 좌측 절반으로 탐색 범위를 좁힘
            end = mid - 1
            continue

        elif seq[mid] < target: # 만약 중간점의 값이 target보다 크다면 우측 절반으로 탐색 범위를 좁힘
            start = mid + 1
            continue

        else: # 중간값이 target과 같은 경우
            if (seq[mid-1] == target) and (seq[mid+1] == target): # 양 옆의 원소 값이 자기 자신과 같은지 확인
                find_end(start=mid+1, end=end, target=target)
                find_end(start=0, end=mid-1, target=target)
            else:
                print(mid)
    return start          
            
N, x = 7, 2
seq = [1, 1, 2, 2, 2, 2, 3]
find_end(start=0, end=len(seq)-1, target=x)
```

```python
"""
Q29 공유기 설치

소요시간: 
메모리:
시간:
https://acmicpc.net/problem/2110

최대한 균등한 간격을 두고 배치가 되어야 최소 거리가 최대가 되리란 것을 짐작할 수 있음. 
만약 공유기가 2개라면 항상 시작, 끝에다 놓는 것이 최적의 결과임. 즉 최대 길이는 전체 array의 끝에서 시작을 뺀 길이와 같음. 

따라서 문제의 조건을 따른다면, 최대 간격은 공유기가 2개일 때이며, 3개 이상일 때부터는 최소 거리를 최대화하려면 무조건 중간점에다가 새로운 공유기를 놔야
거리가 최대화된다는 걸 알 수 있다. 간격이 균등하기 때문에, 역으로 뒤집으면 
간격의 크기 n이 주어졌을 때 이 간격을 유지하면서 c개의 공유기를 균등하게 배치할 수 있느냐 없느냐를 판단해서 해당 간격이 너무 큰 지 작은 지를 파악할 수 있다.

즉 그리디하게 C=3일 때의 간격에서부터 c개의 공유기 배치가 가능한지 여부를 테스트하면서, 책정된 최소거리를 1씩 줄여나가다가 
딱 C개가 배치가 되는 순간 반복문을 멈추면 최대의 최소간격을 보장받을 수 있다. 
==> 1씩 줄이는 건 집의 최대개수가 10억이라서 무조건 안됨. 얘도 이진탐색으로 로그시간 내로 줄여야 함. 

==> 공유기 배치가 가능한지 여부는 그냥 집들을 다 순회하면서 마지막으로 설치한 지점과의 거리가 mid와 같거나 더 클 경우를 세면 됨.
==> 최소 간격을 최대로 하는 간격의 크기는 양끝 사이의 중간값이기 때문. 간격의 크기는 start와 end 사이의 중간값으로 시작함. 
==> 집을 이진탐색하는 것이 아니라 집과 집 사이의 최소 거리를 이진탐색으로 찾는 문제임. 
"""
import sys

n, c = map(int, input().split())
array = list(int(sys.stdin.readline().strip()) for _ in range(n))
array.sort()
start = 0
end = array[-1] - array[0]
result = 0

while (start <= end):
    mid = (start + end) // 2 # mid 계산
    count = 1 # 기본적으로 시작지점에 공유기를 1개 설치함
    prev_loc = array[0]

    for i in range(1, n): # 모든 집들을 순회하면서
        if array[i] - prev_loc >= mid: # 마지막으로 설치한 지점과의 거리가 mid와 같거나 더 클 경우 그 지점에 공유기를 설치할 수 있음. 
            count += 1 # 설치한 공유기 숫자 + 1
            prev_loc = array[i] # 마지막으로 설치한 지점을 갱신
    
    if count >= c: # 순회를 마쳤을 때 설치한 공유기 숫자가 목표치 이상일 경우:
        result = mid  # 설치한 공유기들의 최소 거리는 mid일 것 (mid와 같거나 더 큰 경우 공유기를 설치했으니까)
        start = mid + 1 # c개 이상 설치했을 경우, 공유기 간 거리를 넓혀서 다시 시도해본다. 

    else: 
        end = mid - 1# 만약 c보다 부족할 경우, 공유기 간 거리를 좁힌다.  

print(result) # 가장 마지막으로 c개 설치가 가능한 거리를 result에 저장해놨으므로 탐색이 끝나면 result를 반환한다.
```

```python
"""
Q30. 가사검색

https://school.programmers.co.kr/learn/courses/30/lessons/60060

1. 핵심은 ??? substring이 어느 위치에 있는지랑 상관없는 generalized code를 짜기가 어려우므로, 앞 쪽에 있다는 걸 전제하고 코드를 만든 후,
   ???가 뒤에 있는 거는 아예 뒤집은 채로 저장한 별개의 리스트를 만들어서 동일한 방법을 재사용하는 것. 공간 복잡성을 좀 희생하는 대신 높은 수준의 일반화.

2. 길이에 따라 별개의 리스트로 저장함.

3.  bisect 모듈의 bisect는 리스트 a에서 'x' 원소를 새로 삽입한다면 왼쪽 기준 몇 번 인덱스가 오게 될 것인지 그 인덱스를 반환함. 

reversed dict에서 탐색하는 경우 query도 뒤집어야 함 -> 이걸 간과해서 좀 헤맴
효율성 테스트 3에서 유일하게 시간초과가 남. 시간제한이 많이 빡빡한 듯. 최대한 틈새절약을 하면 될 것 같은데...피곤하니까 여기까지..
"""

from bisect import bisect_left, bisect_right

def count_number(words:list, query:str) -> int:
    """
    우측 극단의 인덱스 - 좌측 극단의 인덱스  
    """
    try:
        left_index = bisect_left(words[len(query)], query.replace('?', 'a')) 
        right_index = bisect_right(words[len(query)], query.replace('?', 'z'))
        return right_index - left_index 
    except KeyError: 
        return 0


def solution(words: list, queries:list) -> list:
    answer = []
    word_dict = dict()
    reversed_word_dict = dict()
    # 단어 길이에 따라 분류해서 dict에 등재
    for word in words:
        length = len(word)
        if length not in word_dict:
            word_dict[length] = [word]
            reversed_word_dict[length] = [''.join(word[::-1])]
        else:
            word_dict[length].append(word)
            reversed_word_dict[length].append(''.join(word[::-1]))
    
    for i in word_dict:
        word_dict[i].sort()
        reversed_word_dict[i].sort()
    
    for query in queries:
        # magic string의 위치 판단 
        if query[0] == '?' and query[-1] == '?': # 맨 앞과 맨 마지막 글자가 둘 다 ?면 전체가 ?로 이루어진 query이다. 이 경우 해당 길이의 word 개수만 반환.
            answer.append(len(word_dict[len(query)]))
            continue

        elif query[0] == '?' and query[-1] != '?': # '?' substr이 앞 쪽에 있을 경우 
            answer.append(count_number(reversed_word_dict, query[::-1]))

        elif query[0] != '?' and query[-1] == '?': # '? substr이 뒤 쪽에 있을 경우
            answer.append(count_number(word_dict, query))       
    return answer
```
