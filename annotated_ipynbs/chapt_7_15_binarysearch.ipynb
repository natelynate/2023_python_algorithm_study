```python
"""
Q27 정렬된 배열에서 특정 수의 개수 구하기 

소요시간: 
메모리:
시간:

솔루션:
이미 상향정렬되어 있으므로 이진탐색을 사용할 수 있음. 값이 x인 원소 n개를 찾아야 하므로, 찾아야 하는 것은 가장 왼편에 있는 x의 인덱스, 그리고 가장 우측에
있는 x의 인덱스를 찾을 수 있으면 서로 인덱스를 빼서 x가 몇 개 있는지 알 수 있다. 

양극단에 위치했는지 여부는 각각 좌측과 우측의 원소 값을 확인해서 어느 한 쪽이 자기자신이 아니라면 = 극단에 위치해 있다고 볼 수 있다. 
"""

#N, x = map(int, input().split())
#seq = list(map(int, input().split()))
N, x = 7, 2
seq = [1, 1, 2, 2, 2, 2, 3]

def find_end(start:int, end:int, target:int) -> int:
    while start < end:
        mid = (end + start) // 2 # 중간점 인덱스 계산
        if seq[mid] > target: # 만약 중간점의 값이 target보다 크다면 좌측 절반으로 탐색 범위를 좁힘
            end = mid - 1
            continue

        elif seq[mid] < target: # 만약 중간점의 값이 target보다 크다면 우측 절반으로 탐색 범위를 좁힘
            start = mid + 1
            continue

        else: # 중간값이 target과 같은 경우
            if (seq[mid-1] == target) and (seq[mid+1] == target): # 양 옆의 원소 값이 자기 자신과 같은지 확인
                find_end(start=mid+1, end=end, target=target)
                find_end(start=0, end=mid-1, target=target)
            else:
                print(mid)
    return start          
            
N, x = 7, 2
seq = [1, 1, 2, 2, 2, 2, 3]
find_end(start=0, end=len(seq)-1, target=x)
```
