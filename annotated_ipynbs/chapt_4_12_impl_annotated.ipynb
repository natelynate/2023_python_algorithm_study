```python
"""
Q07: 럭키 스트레이트

소요시간: 5:05
메모리: 30613KB
시간: 40ms

https://www.acmicpc.net/problem/18406

문제 조건을 순서대로 그대로 구현하면 되기 때문에 자세한 설명은 필요 없음. 
자릿수를 구하기 위해 문자열로 형변환함.
원래 솔루션에 비해 살짝 더 짧아짐
"""

def solution():
    number = str(input())
    length = len(number)
    breakoff = length // 2

    left, right = sum([int(i) for i in number[:breakoff]]), sum([int(i) for i in number[breakoff:]])
    if left == right:
        print("LUCKY")
    else:
        print("READY")
```

```python
"""
Q08: 문자열 재정렬

소요시간 6:35
메모리: nan
시간: nan

입력 문자열을 for문에 넣고 각각 
문자열 클래스의 기본 메서드 isalpha() 혹은 isdigit()을 사용해서 형변화를 직접 하지 않고 숫자/문자를 구분한 후 
별개의 리스트에 넣은 후 따로 sort를 한 결과물을 문자열 접합하면 가능. 
"""

def solution():
    N = input()
    chars = [] # 문자 저장용 리스트
    ints = [] # 숫자 저장용 리스트
    answer = '' # 최종 출력할 문자열

    for character in N:
        if character.isdigit(): # isdigit 메서드가 True를 반환하면 정수로 형변환 후 ints 리스트에 삽입
            ints.append(int(character))

        else:
            chars.append(character) # 그외의 경우 글자이므로 chars 리스트에 삽입
    # 두 리스트를 상향정렬함
    chars = sorted(chars) 
    ints = sorted(ints)

    # 알파벳 순서대로 정렬된 chars 내의 원소들을 빈 문자열 answer에 차례대로 접합
    for i in chars:
        answer += i

    answer += str(sum(ints)) # 정수합을 sum()으로 계산한 후 문자열로 형변환 해서 answer 문자열에 마지막으로 접합
    return answer # 정답 반환

solution()    
```

```python
"""
Q09 문자열 압축

소요시간: 1시간 이상
메모리: nan
시간: nan
https://programmers.co.kr/learn/courses/30/lessons/60057

솔루션:
    '반복되는' 서브문자열과 그 반복횟수를 가지고 원 문자열을 압축하는 것이기 때문에, 이 압축된 서브문자열의 최대 길이는 원 문자열의 1/2에 해당된다는 
    것이 핵심이다. (그 이상부터는 무의미함)
    한 문자열은 1개 이상의 방식으로 압축이 될 수 있기 때문에, 서브 문자열이 가질 수 있는 길이의 범위를 for문으로 탐색하면서 매 iteration마다
    압축된 문자열의 길이를 정수로 보관하고, 그 중 최소인 것을 반환하는 식으로 해결할 수 있다. 

    이중 for문을 만들어서 바깥쪽에서는 서브문자열의 최소 길이부터 최대 길이(1 ~ len(s)//2+1)를 iterate하고, 
    안쪽 반복문에서는 서브문자열를 iterate한다. 이때 prev_slice 변수에 현재 압축이 되는 서브문자열을 저장해놓고, 이후로 들어오는 서브문자열들을 차례대로
    prev_slice와 비교하면서 같으면 count를 증가시키고, 다르면 교체하는 방식으로 서브문자열이 반복되는 횟수를 구해서 압축한다. 
    안쪽 반복문이 끝날 때마다 압축된 문자열의 길이를 저장한다. 
"""
def solution(s): 
    candidates = []
    for i in range(1, len(s) // 2 + 1): # 서브문자열의 최소 길이 1부터 원문자열 길이(s)의 절반 범위를 iterate함
        compressed = '' # 압축문자열
        prev_slice = s[0:i] # 첫 번째 서브문자열은 원 문자열의 0:i 범위이다.
        count = 1 # 카운트 시작점은 1

        for j in range(i, len(s), i):  
            new_slice = s[j:j + i] 
            if prev_slice == new_slice: # 동일한 서브문자열이 반복된다면 count를 1 증가시킴
                count += 1
            else:
                # 압축문자열에 현재까지의 서브문자열 + 개수를 저장함. 만약 1개 밖에 없었다면 숫자 없이 서브 문자열만 저장함
                compressed += str(count) + prev_slice if count >= 2 else prev_slice 
                prev_slice = new_slice # 새로운 서브문자열을 가지고 탐색해야하기 때문에 prev_slice와 count를 갱신함
                count = 1         
        compressed += str(count) + prev_slice if count >= 2 else prev_slice 
        candidates.append(len(compressed)) # 압축된 문자열의 길이를 저장함
    return min(candidates) if candidates else 1
```

```python
"""
Q10 자물쇠와 열쇠
소요시간: 
메모리: nan
시간: nan
https://school.programmers.co.kr/learn/courses/30/lessons/60059

솔루션:
    핵심은 열쇠를 자물쇠에 넣어서 확인할 때 열쇠 전체가 자물쇠 범위 안에 있을 필요는 없다는 것이다. 즉 
    즉 탐색용 for문의 범위 != 자물쇠 크기 이다. 이때 열쇠의 일부분을 그때마다 슬라이싱해서 대입하기엔 너무 번거로우니, 아예 시작하기 전에 입력으로 주어지는
    자물쇠(2차원 리스트)의 크기를 애초에 최대 필요한 크기까지 키워놓고 거기에 열쇠를 넣어가면서 대입하면 된다. 
    
    최대 필요한 크기는 key가 lock에 최소 1칸은 들어가야 한다는 조건 하에 필요한 최대 권역은 key 2개 + lock 1개 - 2 만큼의 길이일 것이다.
    ㅁㅁㅁ  ㅁㅁㅁ
    ㅁㅁㅁ  ㅁㅁㅁ
    ㅁㅁㅅㅅㅅㅁㅁ
        ㅅㅅㅅ
        ㅅㅅㅅ

    key와 lock이 모두 3x3이라는 가정 하에 필요한 최대 넓이는 위 그림처럼 7x7이다 = (len(key) * 2) + len(lock) - 2
    탐색 범위는 한 위치에서 90도 회전을 3번할 수 있으므로 전체 for문에서 반복하게 되는 iteration 개수는 전체 판의 크기 *3가 된다. 
    key와 lock은 최대 20이므로 58 * 3 = 174번이 최대가 된다. 즉 완전탐색이긴 하지만 반복횟수 자체가 적으므로 안전할 것. 

    사전준비를 제외하면 알고리즘 자체는 굉장히 간단한데, 그냥 이중 for문으로 key를 특정 위치에 대입해보고, 만약에 일치하면 
    조기종료를 하고, 그게 아니라면 4방향 회전 -> 그 후 자리 이동을 전체 확장된 lock 범위를 돌면서 반복하면 끝이다.
"""
import copy

key = [[0, 0, 0], [1, 0, 0], [0, 1, 1]]
lock = [[1, 1, 1], [1, 1, 0], [1, 0, 1]]

def solution(key, lock)-> bool:
    def rotate_key(key):
        """
        열과 행을 뒤바꾼 후 reverse를 적용해 2차원 리스트를 시계방향으로 90도 회전함
        """
        rotated_key = []
        for row in range(len(key)):
            rotated_key.append([i for i in reversed([key[col][row] for col in range(len(key))])])
        return rotated_key


    def insert_key(upperleft_row, upperleft_col, key, lock, lock_map)-> bool:
        """
        (row, col)이 키의 좌측 최상단 좌표일 때 전달받은 임의의 key를 lock_map의 값과 합쳤을 때 중앙의 original lock 부분의 합이 모두 1인지 확인함
        """
        # 
        map_ = copy.deepcopy(lock_map)
        for i in range(len(key)):
            for j in range(len(key)):
                map_[i + upperleft_col][j + upperleft_row] += key[i][j]
        
        print(f"key: with upperleft at {col},{row}")
        for i in key:
            print(i)
        
        for i in map_:
            print(i)

        print("====")
        
        # 중앙 부분의 lock 범위 내의 타일에 1이 아닌 값이 하나라도 있을 경우 False 반환
        for i in range(len(lock)):
            for j in range(len(lock)):
                if map_[i + len(key) - 1][j + len(key) -1] != 1:
                    return False
        # 모두 통과하면 True 반환
        return True

            
    # lock의 크기를 모든 key의 경우의 수를 지원할 수 있는 한도로 키워서 lock_map이란 이름으로 따로 저장함. 빈 칸은 0으로 채움
    map_size = (len(key) * 2) + len(lock) - 2
    lock_map = [[0] * map_size for _ in range(map_size)]


    # lock_map의 중앙에는 원래 lock값으로 채움
    for col in range(len(lock)):
        for row in range(len(lock)):
            lock_map[col + len(key) - 1][row + len(key) - 1] = lock[col][row]

    for i in lock_map:
        print(i)

    # 탐색 범위는 
    for col in range(map_size - len(key) + 1):
        for row in range(map_size - len(key) + 1):
            for dir in range(4):
                key = rotate_key(key)
                fit = insert_key(col, row, key, lock, lock_map)
                if fit:
                    return True
    return False

solution(key, lock)
```

```python
"""
뱀 정보를 클래스로 구현하였고 
시간을 defaultdict에 넣어서 매 iteration마다 현재 time을 key로 defaultdict를 검색한다. 만약 기존에 저장된 정보가 없으면 방향 전환이 필요없는 시간이라는 
뜻이므로 스킵하고, 만약 시간을 key로 하여 검색했을 때 방향전환 정보가 들어있는 시간이었다면 그대로 방향을 바꾸는 방식으로 동작한다. 
코드가 간편하나, defaultdict가 굉장히 sparse한 형태로 공간을 많이 소요한다는 단점이 있음. 

"""

from collections import defaultdict, deque

class baem:
    def __init__(self):
        self.pos = deque([[1, 1]]) # 뱀이 위치한 타일 좌표들을 queue로 저장함
        self.dir = 2 # 기본 방향


    def move_forward(self, dx, dy):
        """
        현재 바라보고 있는 방향으로 1칸 전진 후 도착한 좌표를 반환하고 queue에서 머리 위치를 갱신함
        """
        x = self.pos[0][1] + dx[self.dir]
        y = self.pos[0][0] + dy[self.dir]
        self.pos.appendleft([y, x])
        return x, y


    def change_dir(self, event_dict, time):
        """
        방향 90도로 왼/오른쪽으로 전환
        """
        dir = event_dict[time]

        if dir == 'L': # 왼쪽
            self.dir = self.dir - 1 if self.dir > 0 else 3
            
        elif dir == 'D': # 오른쪽
            self.dir = self.dir + 1 if self.dir < 3 else 0


def solution():
    N = int(input())
    K = int(input())
    
    # Render world map
    game_map = [[0] * (N + 2) for _ in range(N + 2)]

    # 외곽 타일을 6으로 지정해서 배치
    for tile in range(N + 2):
        game_map[0][tile] = 6
        game_map[-1][tile] = 6
        game_map[tile][0] = 6
        game_map[tile][-1] = 6
    # 좌측 최상단에 뱀을 정수 2로 표시
    game_map[1][1] = 2
    
    for apple in range(K):
        # Put apples 
        y, x = input().split()
        game_map[int(y)][int(x)] = 1 

    # 시계방향으로 왼쪽, 위, 오른쪽, 아래
    dx = [-1, 0, 1, 0] 
    dy = [0, -1, 0, 1]
   

    # 방향전환 정보를 dictionary에 저장
    event_dict = defaultdict(int) # default value를 0으로 책정
    P = int(input()) # 방향전환 횟수 
    for i in range(P):
        info = input().split()
        event_dict[int(info[0])] = info[1]

    time = 0
    snake = baem() # baem 인스턴스 생성
    while True:
        time += 1 # Count time in sec
        print(f"{time}초")
        for i in game_map:
            print(i)
        print("=" * 10)

        # 현재 방향으로 1칸 전진
        x, y = snake.move_forward(dx, dy)
        print(f"moved to- {y},{x}")  
        
        # 도착한 타일에 뱀 몸이 있거나 혹은 맵 외곽으로 나갈 경우 소요시간을 반환하고 종료
        if game_map[y][x] == 6 or game_map[y][x] == 2:
            break
        
        # 도착한 타일에 사과가 있으면 도착한 타일의 좌표로 뱀 머리 좌표를 갱신하고 맵 타일의 값도 2로 변경
        elif game_map[y][x] == 1:
            game_map[y][x] = 2

        # 도착한 타일에 사과도 없고 외곽도 아니며 뱀 몸통도 아니라면 그냥 전진
        else:
            game_map[y][x] = 2
            tail_y, tail_x = snake.pos.pop() # pop이 기본적으로 맨 오른쪽 끝부터 pop하므로 이걸 통해 꼬리 좌표를 불러옴
            game_map[tail_y][tail_x] = 0

        # X초가 끝난 후 방향을 바꿈. 
        # 만약 방향 변환이 있다면 방향을 전환함. 왼쪽이면 -1, 오른쪽이면 +1, 범위를 벗어나는 경우를 위해 if문으로 확인해서 맞는 값을 재할당해준다
        snake.change_dir(event_dict, time)

    print(time)
```

```python
"""
Q12: 기둥과 보 설치


소요시간: 3h+
메모리: 
시간:
https://school.programmers.co.kr/learn/courses/30/lessons/60061 

솔루션:
    전체 격자가 크기가 최대 100, 그리고 구조물의 최대 숫자가 1000이고, 구조물 규칙 때문에 모든 칸들이 다 사용되지 못하리라고 가정하면 전체 데이터 수 자체가
    작기 때문에 구조물 조건을 만족하는지에 대한 여부를 매 iteration마다 전체 구조물 리스트를 iterate하면서 하나씩 다 확인하는 다소 비효율적인 방식을 써도
    통과할 수 있다. 

    struct_list에 현재까지 설치된 보와 기둥 오브젝트들을 튜플로 다 보관하고 있다가, 설치든 삭제든 어떠한 작업을 완수하고 날 때마다 전체 오브젝트들을 
    for문으로 iterate하면서 구조물 종류에 따라 건축 조건을 만족하는지 확인하고 만족하는 경우엔 통과하고 그렇지 않으면 바로 직전에 수행한 작업을 
    그냥 반대 작업을 수행하는 방식으로 롤백해버리면 된다(즉 수행한 게 삭제였으면 동일한 위치에 구조물을 추가하고, 추가 작업이었으면 설치했던 걸 삭제)

그 외:
 # for문 내에서 dictionary 내 content를 iterate할 때 content를 modify하면 
 # RuntimeError: dictionary changed size during iteration 오류 발생할 수 있음. .items()를 사용하면 방지할 수 있다. 
"""

def check_integrity(struct_list):
    for struct in struct_list:
        x = struct[0]
        y = struct[1]
        if struct[2] == 0:
            if y == 0: # 기둥은 해당 좌표의 y수치가 0이거나(바닥)
                continue
            if (x-1, y, 1) in struct_list or (x, y, 1) in struct_list: # 왼편에 1이 있거나(보 위에 있음) 설치 지점에 1이 있을 경우 통과
                continue
            if (x, y-1, 0) in struct_list: # (x, y-1)에 0이 있을 경우 (다른 기둥 위에 있음) 통과
                continue
            return False

        elif struct[2] == 1: # 보의 경우
            if (x, y-1, 0) in struct_list: # 바로 아래에 기둥이 있을 경우 통과
                continue
            if (x+1, y-1, 0) in struct_list: # 우측 하단에 기둥이 있을 경우 통과
                continue
            if (x-1, y, 1) in struct_list and (x+1, y, 1) in struct_list: # 양 옆에 보가 있을 경우 통과
                continue
            return False
    return True


def solution(n, build_frame):
    struct_list = list()
    for workorder in build_frame:
        # struct_type = 구조물 종류(기둥/보), work_type = 작업 종류(추가/삭제)
        x, y, struct_type, work_type = workorder
        if work_type == 1:
            struct_list.append((x, y, struct_type)) # 구조물 추가
            if not check_integrity(struct_list): # 위반하는 규칙이 없을 경우 continue
                struct_list.pop() # 규칙에 위반되면 다시 롤백. pop removes the item inserted last
                print(f"roll back add item {x},{y},{struct_type},{work_type}")
            else:
                print(f"pass add request {x},{y},{struct_type},{work_type}")
                
        
        # 작업 종류가 '삭제'일 경우 '추가'와 정확히 반대로 동일한 프로세스 진행
        elif work_type == 0: 
            struct_list.remove((x, y, struct_type))
            if not check_integrity(struct_list):
                struct_list.append((x, y, struct_type)) # 구조물 추가 
                print(f"roll back delete item {x},{y},{struct_type},{work_type}")
            else:
                print(f"pass delete request {x},{y},{struct_type},{work_type}")
    struct_list.sort()
    return struct_list
```

```python
'''
Q13: 치킨 배달
풀이 시간: 40분
제한: 1초, 512mb

유사코드:
해결: 
메모리: 
시간:
https://www.acmic.net/problem/15686

솔루션:
    도시 크기 N과 치킨집 개수 M이 각각 최대 50, 13으로 매우 작으므로, 완전탐색으로 풀기로 결정함.
    최대 13개에서 M개를 고르게 되므로, 13Cm 중 m이 무엇이 되더라도 10만을 넘기지 않는다. 
    최대크기는 13C6 or 13C7인데 1716개이므로 집 개수가 100개 이하이므로 1716 * 100 = 171600개가 최악의 경우가 된다.
    1초 제한시간의 연산 상한선은 2000만번이므로 여유가 있는 편이다. 

    완전탐색을 위해 itertools의 combinations를 사용했고, 이중 for문(치킨집 조합-> 집들)을 사용해서 각 경우의 최소 치킨거리를 모두 계산해놓고 
    별도의 리스트에 저장했다가 min()으로 그 중 최소값을 반환하는 방식을 취함.  
'''

import itertools
from itertools import combinations

# 집과 치킨집의 맨하탄 거리 계산
def calculate_distance(house, chicken):
    return abs(house[0] - chicken[0]) + abs(house[1] - chicken[1])

chickens = [] # 치킨집 좌표 저장 리스트
houses = [] # 집 좌표 저장 리스트
city_chicken_distances = [] # 
chicken_distances = []

N, M = map(int, input().split())

for row_num in range(1, N+1):
    row = [int(k) for k in input().split()]   # 도시 정보 입력
    for index, value in enumerate(row): 
        if value == 1:
            houses.append([row_num, index+1]) # 집 리스트에 좌표 추가
        if value == 2:
            chickens.append([row_num, index+1]) # 치킨집 리스트에 좌표 추가

possible_combinations = list(itertools.combinations(chickens, M)) # 선택할 수 있는 치킨집 조합 계산

for combination in possible_combinations: # 각 조합별로 치킨 거리를 계산함
    for house in houses: # 각 집마다 치킨 거리를 계산하고 최소 치킨거리를 찾음
        candidates = [calculate_distance(house, chicken) for chicken in combination] # 특정 집 1개와 다른 치킨집들 사이의 모든 거리 
        candidates = min(candidates) # 해당 집의 최단 치킨 거리
        chicken_distances.append(candidates) # 집의 최단 치킨 거리를 chicken distances에 append
    
    # 모든 집에 대해 최단 치킨 거리를 계산하고 나면 그것들의 합이 해당 combination의 도시 치킨 거리이다
    sum_chicken_distances = sum(chicken_distances)
    city_chicken_distances.append(sum_chicken_distances)
    chicken_distances = []  # 초기화

print(min(city_chicken_distances)) # 모든 조합들 중 최단 치킨거리를 반환
```

```python
"""
Q14 외벽점검

소요시간: ?
메모리: ?
시간: ?

https://school.programmers.co.kr/learn/courses/30/lessons/60062
해설 코드 참조
"""

from itertools import permutations

def solution(n, weak, dist):
    """
    <해설 참조> 

    친구가 최대 8명이기 때문에 범위가 좁아서 permutation으로 가능한 모든 조합을 테스트하는 방식으로도 풀 수 있음.
    8명 내에서 가능한 최소 조합 (1명)부터 8명까지 모든 경우의 수를 다 순열로 계산해서 그때마다 모든 취약점이 커버될 때까지 걸리는 시간을
    저장하도록 하고 그 중 최소 시간을 반환하면 된다.
    
    순열을 계산하면 투입될 친구들이 리스트로 생성되고, 여기서 순서대로 투입해서 한 instance마다 점검할 수 있는 취약점들을 점검하도록 하면 된다. 

    positions는 특정 취약점에서 특정 친구가 탐색을 시작했을 때 탐색할 수 있는 거리이고 (취약점 위치 + 친구 이동거리)
    각 친구가 탐색할 때마다 각 취약점이 현재 positions보다 작으면 점검될 수 있으므로 패스. (오름차순으로 정렬되어 있는) 취약점들 중 하나가 현재 positions
    보다 크면 (현재 친구가 탐색할 수 있는 거리보다 크다면) 해당 취약점부터 다음 친구가 탐색을 해야 하므로 positions를 새로 계산하고, 투입되는 친구의
    숫자를 표시하는 count를 증가시킨다. 만약 count가 iteration중 전체 친구의 수보다 커지게 되면 break하고 최종적으로 필요한 친구 수를 반환한다. 

    
    # 무조건 시계방향으로 간다고 가정하는데, 반시계-시계 두 방향 모두 갈 수 있다는 특징은 순수하게 문제 푸는 사람 헷갈리게 하려는 트릭인가...?
    -> 생각해보니 맞음. 어차피 무조건 모든 친구들은 특정 취약점 위에서부터 시작한다고 가정하면 매번 투입할 때마다 최소 한 개의 취약점은 점검할 수 있음.
    그럼 취약점이 상향정렬되어 있으면 한 방향으로만 찾으면 됨. (만약 반대 방향에는 직전 친구가 점검했을 테니까 반대 방향에는 취약점이 '무조건' 없기 때문에
    내가 시계방향으로 이런 방식으로 탐색하고 있었으면 절대 시계반대방향을 고려할 필요가 1도 없음) 


    # 원형으로 되어 있는 탐색범위라서 코딩하기 헷갈릴 때는 그냥 2배로 리스트를 잡아늘려서 일자로 변환해서 탐색할 수 있다.
    """
    length = len(weak) # 취약점 개수 
    for i in range(length): 
        weak.append(weak[i] + n) # 리스트를 두 배로 늘려서 기존의 원형 리스트를 일자로 표현
    answer = len(dist) + 1 # 답은 최대 친구 수 + 1로 초기에 책정
    

    for start in range(length): # 투입된 친구 수
        for friends in list(permutations(dist, len(dist))): # 투입된 수 만큼 친구들의 순열 생성해서 iterate함.
            count = 1  # 현재 동원된 친구 수 # 무조건 1명은 시작부터 들어가고 
            position = weak[start] + friends[count - 1] # 각 취약점에서부터 탐색을 시작한다고 가정
            for index in range(start, start + length): # 현재 취약점에서부터 탐색 범위 내를 iterate함
                if position < weak[index]: # 만약 최대 탐색 범위보다 취약점이 더 멀리 있다면 친구가 하나 더 필요함
                    count += 1
                    if count > len(dist): # 투입되어야 할 인원 수가 전체 친구수보다 많아지면 종료함
                        break
                    position = weak[index] + friends[count - 1] # 최대 탐색 범위 갱신
            answer = min(answer, count)

    if answer > len(dist): # 만약 전체 친구 수보다 투입되어야 하는 인원이 많다면 -1 반환
        return -1 

    return answer
```
