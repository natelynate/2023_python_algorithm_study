```python
"""
Q23 국영수

소요시간: 16:39 ~16:55 (15분)
메모리:
시간:
https://www.acmicpc.net/problem/10825

이름, 국어, 영어, 수학 순서이므로 
key=lambda x: -x[1], x[2], -x[3], x[0] 

그냥 sorted함수에서 key argument에 람다함수를 집어넣는 방법을 알면 바로 풀 수 있음
"""
import sys

# N = int(input())
# data = [input().split() for i in range(N)]
data = [sys.stdin.readline().split() for _ in range(int(input()))]

data = [(i[0], int(i[1]), int(i[2]), int(i[3])) for i in data]
data = sorted(data, key=lambda x: (-x[1], x[2], -x[3], x[0]))
for i in data:
    print(i[0])
```

```python
"""
Q24 안테나

소요시간: 17:00~
메모리:
시간: 132
https://www.acmicpc.net/problem/18310

솔루션: 어렵게 생각했으나 그냥 전체 리스트에서 중간값에 해당하는 원소를 반환하면 되는 문제였음
"""
# <그냥 중간값 반환하면 된다 ㅡㅡ>

N = int(input())
houses = list(map(int, input().split())) # list(map(int, sys.stdin.readline().split()))
houses.sort()
print(houses[((N-1)//2)])
```

```python
"""
Q25 실패율

소요시간: 12:19~
메모리:
시간:
https://school.programmers.co.kr/learn/courses/30/lessons/42889

솔루션:

총 3개의 array를 선언한다. 세 array 모두 index로 stage번호를 나타낸다.
stuck array에서는 각 스테이지에 현재 도달해있는 플레이어들의 배열이다 (인덱스 편의성을 위해 제공되는 리스트에서 크기만 키움)
passed array는 각 스테이지별로 통과한 사람의 수이다. N번 스테이지에 있는 사람은 1번, 2번, ... N-1번까지를 무조건 통과했다는 뜻이니 
(예컨대 3번 스테이지를 하고 있는 사람이 3명이면 1번과 2번을 통과한 사람도 최소 3명이라는 것)
이중 for문으로 이를 계산해서 각 스테이지별로 통과한 사람을 원소로 저장함. 

failrate array에서는 각 스테이지별로 failrate를 계산하는데 그냥 stuck과 passed array에서 동일한 인덱스의 원소를 나누면 바로 실패율이 나온다. 
이를 failrate의 순으로 하향정렬하고 이름만 따로 빼내서 출력하면 된다.
"""

def solution(N, stages):
    stuck = [0] * (N + 2) # 해당 인덱스의 스테이지에서 막힌 사람들
    passed = [0] * (N + 2) # 해당 인덱스의 스테이지를 지나간 사람들의 총합
    fail_rates = []

    for stuck_stage in stages:  
        stuck[stuck_stage] += 1
        for i in range(1, stuck_stage+1):
            passed[i] += 1
    
    else:
        # (스테이지번호, 스테이지 실패율)의 튜플 형태로 저장함
        for i in range(1, N+1):
            if passed[i] == 0:
                fail_rate = (i, 0) # 지나간 사람이 아무도 없으면 실패율은 0
                fail_rates.append(fail_rate) 
            else:
                fail_rate = (i, stuck[i] / passed[i]) # 못 지나간 사람 / 지나간 사람 으로 실패율을 구함
                fail_rates.append(fail_rate)

    fail_rates = sorted(fail_rates, key=lambda x: x[1], reverse=True) # 1번 인덱스 기준으로 하향정렬
    result = [i[0] for i in fail_rates] # 리스트 컴프리헨션으로 번호만 남김
    return result

solution(4, [1,1,1,1,1])
```

```python
"""
Q26 카드 정렬하기

소요시간: 15:15
메모리:
시간: 3972
https://www.acmicpc.net/problem/1715

그리디하게 현재 카드를 상향정렬한 상태에서 가장 작은 카드들끼리 (왼쪽에서부터) 한 쌍씩 묶는 것이 합을 가장 작게 한다. 
N <= 100,000 이고 카드더미 크기는 <= 1000 이므로
최대값은 1000 * 100,000 = 1억이다

무조건 두 쌍씩 묶는 것이 능사가 아님. 10 10 10 40 같은 경우면 10+10 하고 20+10 한 후에 30+40 하는 게 처음부터 두 쌍씩 묶는 것보다 효율이 좋음.
즉 맨 앞의 인덱스 0, 1을 묶고, 그 결과를 다시 정렬하고, 그걸 전체에서 1개가 남을 때까지 반복해야함. 
즉 매번 정렬됨이 보장되어야 하는데 그걸 직접 삽입정렬이나, 기본 sort를 할 경우엔 시간초과를 피할 수 없음. 그러므로 명시적으로 정렬할 필요가 없는 
우선순위 queue를 사용해야 시간 안에 완료되게 할 수 있다. 

우선순위 queue는 원소가 출입할 때마다 자동으로 정렬되므로 우선순위 queue를 heapq라이브러리로 생성하는 것이 제일 확실한 방법임
"""
import heapq
from heapq import heapify

def solution():
    N = int(input())
    if N == 1: # 카드 개수가 1개면 0 반환
        return 0
    heap = [int(input()) for _ in range(N)]
    heapify(heap) # 우선순위 queue 생성
    result = 0

    # heapq에 원소가 1개 남을 때까지 반복
    while len(heap) != 1:
        new_deck = heapq.heappop(heap) + heapq.heappop(heap) # 선출되는 원소 2개를 합한 값을 result에 더한 후 그 합을 heapq에 새로 push함
        result += new_deck
        heapq.heappush(heap, new_deck)
    return result 

solution()
```
