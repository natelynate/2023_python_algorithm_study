```python
"""
Q01 모험가 길드
소요시간: 1h
메모리: 
시간: 

기출: 핵심유형
솔루션:
    가장 많은 모험가들을 보내야 하므로, 공포 내성이 강한 사람들일 수록 전부 다 내보내야 최적해를 찾을 공산이 크다. 그러려면 상향정렬을 해서 숫자가 낮으면 바로 보내면 된다. 
    1일 경우 혼자만으로도 그룹 결성이 되므로 무조건 내보내진다. 
    
    상향정렬을 수행한 후 아래서부터 만들어지는 대로 내보내다 보면 (모두 다 내보낼 수 없다는 가정 하에) 어느 시점에는 더 이상 내보낼 수 없는 상태가 될 텐데, 
    새로운 사람을 내보낼 수 '없는' 경우는 현재 새로운 사람의 공포 내성보다 현재 그룹 총원 숫자가 더 작을 경우이다. 
    즉 상술한 이유에 해당하는 경우라면 다음 사람을 추가해서 정원을 확대해야 하고, 만일 더 내보낼 사람이 없다면 종료해야 할 것이다. 

    필요한 오브젝트는: 
        1. 전체 모험가들의 공포 내성을 상향정렬한 array
        2. 내보낼 수 있는 최대 그룹 숫자를 저장하는 변수 answer
        3. 시행 때마다 0으로 초기화되는, 현재 그룹 크기를 저장하는 변수 group_size
        
    를 사용한다.     
    

"""
test_cases = [[2, 3, 3, 5, 6],
              [1, 1, 3, 3, 5, 6],
              [5, 5, 5, 5, 5, 5],
              [1, 1, 1, 1, 1, 1],
              [1, 2, 2, 3, 3, 4, 4, 4, 5, 5, 5, 5, 6],
              [4, 1, 1, 1, 1, 1, 3]]

test_case_answers = [1, 2, 1, 6, 4, 5]


def solution(array):
    array.sort() # sort함수로 상향정렬함
    answer = 0 
    group_size = 0

    for fear_level in array: 
        group_size += 1 # 그룹 크기를 1 증가
        if fear_level > group_size: # 만약 현재 보고 있는 사람의 공포도가 현재 그룹크기보다 크다면 출정할 수 없음
            continue
        else: # 그 외의 경우가 출정이 가능하므로 
            answer += 1 # 출정 가능한 파티 숫자를 1증가
            group_size = 0 # 다음 파티 결성을 위해 group_size 변수 초기화
    
    return answer # 정답 반환


answers = [solution(case) for case in test_cases]
if answers == test_case_answers:
    print(True)
```

```python
"""
Q03: 문자열 뒤집기
https://www.acmicpc.net/problem/1439

소요시간: 문자열 따라서 iterative하게 scan해서 쪼개는 방식 고집했다가 좀 더 오래걸린 듯 한 1시간 반..?
메모리: 
시간: solution1: 44ms
      solution2: 40ms
솔루션:
    0과 1로 구성된 문자열 중에서, 최소한의 뒤집기를 통해 0혹은 1로만 이루어진 문자열로 변환해야 하므로, 결국 0과 1중 더 개수가 적은 쪽을 뒤집어야 한다는 말과 같다. 
    즉 주어진 문자열에서 가장 적은 sub문자열의 개수를 반환하는 문제로 치환될 수 있다. 
    
"""

def solution1():
    """
    몇 개의 sub문자열로 이루어져 있는지 카운트할 때 좌측 극단과 우측 극단의 인덱스를 관리하면서 좌측 극단 인덱스 번째의 값과 우측 극단 인덱스 번째의 값이 달라지면 sub배열이 달라졌음을 
    감지하는 방식. sub배열이 달라졌으면 그때까지 카운트해왔던 배열을 저장함. 
    """
    S = input() 

    seq_dict = {'0':[], '1':[]} # 숫자 sub배열을 저장할 dictionary
    left_end = 0 # 좌측 극단 인덱스
    right_end = 0 # 우측 극단 인덱스

    while right_end < len(S): # 우측 극단이 전체 길이를 벗어나지 않는 선에서 전체를 반복함
        count = 0 # sub문자열 길이
        for index, value in enumerate(S[left_end:]): # 좌측 인덱스 극단부터 중단되기 전까지 반복
            if S[left_end] == S[right_end]:
                count += 1
                right_end += 1
                continue
            seq_dict[S[left_end]].append(count) # count 저장
            left_end = right_end # 위치 옮기고
            break

        else: ## 필요한지 한 번 더 확인
            seq_dict[S[left_end]].append(count)

    print(len(seq_dict['0']) if len(seq_dict['0']) < len(seq_dict['1']) else len(seq_dict['1']))

    
def solution2():
    """
    복잡하게 좌우측 극단 인덱스들을 관리할 필요 없이 for문으로 보면서 매 instance가 달라졌는지 안 달라졌는지만 감시하는 방식. 
    교재 해답과 같음. 
    """
    S = input()
    seq_dict = {'0':0, '1':0}
    ones = 0 # 0으로 이루어진 substring 개수 저장
    zeros = 0 # 1으로 이루어진 substring 개수 저장
    
    prev = S[0] #
    for current in S[1:]:
        if prev == current: # 숫자가 바뀌지 않으면 넘어감
            continue
        else:
            seq_dict[prev] += 1 # 숫자가 바뀌었으면 prev에 저장되어 있었던 숫자(0/1)에 해당되는 value를 증가시킴
            prev = current # prev를 현재 보고 있는 값으로 지정해줌

    else: # 맨 마지막 substring의 경우 개수 증가 조건인 '다음 숫자가 현재 숫자와 같지 않음'에 해당되지 않으므로 자동으로 카운트가 안되기 때문에 수동으로 증가시킴.
        seq_dict[current] += 1 # 가장 마지막 숫자에 해당하는 value를 증가시키면 됨
    
    print(min(seq_dict['0'], seq_dict['1'])) # 0과 1로 이루어진 substring개수 중 적은 쪽을 반환함
```

```python
"""
Q04: 만들 수 없는 금액
소요시간:
메모리: 
시간: 
솔루션:
    만들 수 없는 양의 금액 중 최솟값이므로, 작은 금액에서부터 시작해서 해당 금액을 만들 수 있는지에 대한 여부를 검증하면서 만들 수 있으면 상한선을 늘리고 또 재검증하는 방식으로 진행하다가 
    더이상 만들 수 없는 첫 번째 경우를 만나면 거기서 반복을 멈추고 값을 반환하는 방식으로 설계한다.
    
    만들 수 있는 금액을 초기에 1에서 시작하고, 해당 가치를 만들 수 있다고 검증되면(즉 새로운 동전의 액면가가 목표 금액보다 같거나 작다면) 그 새로운 동전의 액면가를 목표 가치에 더하므로
    매번 시행때마다 "만들 수 있다고 검증된 가치 + 1" 의 가치를 만들 수 있는 지 매번 검증하는 식이 된다. 그래서 x번째 시행에서 이 가치를 만들 수 없다고 나오게 되면 바로 직전까지의 가치를 
    반환하면 그것이 만들 수 없는 금액 중 최솟값일 것이다.

    매번 새로운 동전의 액면가가 목표 금액보다 같거나 작은지를 보는 이유는 만약 새로운 동전이 목표 액수보다 커 버리면 죽었다 깨어나도 그 새로운 동전으로는 목표 액수를 만들 수 없고, 
    동시에 바로 직전 시행에서 이전 동전들로는 만들 수 없다는 것이 증명되었으므로(기존합에서 +1한 것이 목표 액수니까) 목표 금액을 만들려면 무조건 새로운 동전은 해당 목표 금액과 같아서 
    단독으로 그 금액을 만들거나, 혹은 그 금액보다 작아서 이전 동전들과의 합으로 그 금액을 만들어야 하기 때문이다. 

"""

N = int(input())
data = list(map(int, input().split()))
data.sort() # 동전 가치를 상향정렬함

target = 1 # 현재 만들 수 있는 금액
for coin in data: 
    if coin > target: # 현재 동전 액면가가 목표 금액보다 높을 경우 
        break
    else:
        target += coin

print(target)
```

```python
"""
Q05: 볼링공 고르기
소요시간:
메모리: 
시간: 
솔루션:
    A가 특정 무게의 볼링공을 선택했을 때, B가 선택할 수 있는 다른 공들의 개수를 합하는 것을 모든 무게들에 대해서 반복한 총합이 답이다. 
    편의를 위해 미리 각 무게 별로 볼링공이 몇 개 있는지 세고 진행한다(미리 하지 않고 나중에 반복문 때 가서 하면 (길이 x 무게 종류)번 반복해야 하기 때문에 비효율적이다.
    
    여기서 이미 계산했던 조합은 제외해야 하기 때문에, B가 택하는 볼링공은 항상 A가 택한 것보다 무거운 것만을 합산한다 (왜냐하면 반대의 경우는 이전 시행에서 이미 합산되었을 것이므로)

"""
n, m = map(int, input().split())
data = list(map(int, input().split()))
answer = 0
per_weight = [0 for _ in range(m + 1)] # 최대 무게만큼 0으로 채워진 리스트를 선언. 인덱싱 편의를 위해 +1 증가시킴
for i in data:
    per_weight[i] += 1 # 인덱스는 무게를 나타내고, 리스트의 값은 해당 무게의 볼링공의 개수를 나타냄

for i in range(m): # 무게 종류만큼 반복함
    answer += sum(per_weight[i+1:]) * per_weight[i] # A가 택한 무게보다 무거운 볼링공들의 무게 합 x A가 택한 무게의 볼링공 개수
    print(answer)
```

```python
"""
Q06: 무지의 먹방 라이브

소요시간: 2일(ㅋㅋ)
메모리: 
시간: 
https://school.programmers.co.kr/learn/courses/30/lessons/42891

solution1 ~ solution4는 직접 풀었는데 계속해서 나아지지만 효율성 테스트를 전부 통과 못함. 
solution5는 공식 해답 참조. 개별 solution에 대한 분석은 각 함수의 코멘트 참조.
"""
def solution(food_times, k):
    """
    처음에는 반복문을 접시 단위로 반복하도록 설계했지만 이럴 경우 효율성 테스트를 통과 못함
    """
    pos = 0
    time = 0
    platter_length = len(food_times) - 1 # 음식 개수
    print(food_times, f"pos={pos}, time={time}")
    while time < k: # 네트워크 장애가 발생할 때까지 반복
        if food_times[pos] != 0: # pos번째 음식이 아직 남아 있으면 1초 동안 그 음식을 먹음
            food_times[pos] -= 1
            time += 1

            # pos 위치 갱신함. 만약 pos가 끝까지 오게 되면 다시 첫번째 (0번 인덱스)로 회전
            if pos < platter_length:
                pos += 1
            else:
                pos = 0
        
        else: # 해당 pos에 음식이 없는 경우
            if pos < platter_length:
                pos += 1
            else:
                pos = 0
        print(food_times, f"next_pos={pos}, time={time}")
        
    # k초 후 네트워크 장애 발생하면 현재 pos에서 가장 가까운 음식의 인덱스+1를 반환함
    while True:
        if sum(food_times) == 0: # 음식이 하나도 없으면 -1 반환
            return -1

        if food_times[pos] != 0:
            return pos + 1
        else:
            if pos < platter_length:
                pos += 1
            else:
                pos = 0


def solution2(food_times, k):
    from collections import deque
    """
    iteration 단위를 개별 food가 아니라 food_times 중 min_value로 하자
    매 iteration 마다 무조건 현재 남아있는 음식들을 고려했을 때 최대한 많은 왕복 수를 거침 -> Greedy
    마지막 남은 시간만큼 1개 단위로 queue처럼 iterate 하는 것도 queue를 쓰든 리스트로 하든 똑같이 통과는 못함 ;
    앞 쪽에서 runtime만큼 빼는 걸 더 많이 뺄 수 있게 해야 할 듯
    단순히 최소 음식 x 길이 가 k보다 커야 사전에 뺄 수 있게 하면 너무 제한적이라서 full iteration이 빈번하게 발생함. 
    """
    if sum(food_times) < k:
        return -1

    food_times = sorted([(value, index) for index, value in enumerate(food_times)], reverse=True) # food_times를 
    while True: # 서비스 장애가 발생하기 전까지 반복
        # 가장 적은 음식량 min_food가 0이 될 때까지 완주를 반복했을 때 걸리는 시간 = 최소 음식량 x 음식 개수 second가 소요됨
        if not food_times: # 중간에 food_times가 소진되면 -1반환 후 종료
            return -1

        runtime = (len_food := len(food_times)) * (min_food := food_times[-1][0]) 

        if k - runtime < 0: # 만약 runtime을 추가했을 때 오류 발생시간을 넘거나 while문 종료
            break

        # 만약 시간 여유가 있으면 runtime 만큼 먹방을 진행.
        k -= runtime # runtime 만큼 시간을 빼주고
        food_times = [[i[0] - min_food, i[1]] for i in food_times if i[0] != min_food] # 이 경우 min_food만큼 남았던 음식은 다 먹어서 0이 될 터이니 전체 리스트에서 뺀다
        print(food_times, k)
    
    # while문이 끝나면 현재 남은 k초 만큼 더 전진하고, 장애가 발생하면 장애가 발생한 시점 이후에서 가장 가까운 음식의 인덱스를 반환하는데, 
    # 이는 실질적으로 k + 1초 만큼 전진해서 멈춘 그 자리의 인덱스를 반환하라는 것과 동치함. 
    print(f"remaining time={k}, {sorted(food_times, key=lambda x: x[1])}")

    food_times = deque([i for i in sorted(food_times, key=lambda x: x[1])])
    pos = 0
    while k >= 0:
        if not food_times:
            return - 1
        current_dish = food_times.popleft()
        k -= 1
        if current_dish[0] - 1 > 0: 
            food_times.append([current_dish[0]-1, current_dish[1]])
    else:
        return current_dish[1] + 1


def solution3(food_times, k):
    from collections import deque
    """
    solution2에 비해 시작할 때 더 빼고 시작하는데...아마 while True하고 k를 미리 줄일 때 food_times 리스트를 계속 초기화하는 것이 비효율적이라 그럴 듯?

    얘도 효율성 테스트를 1개만 통과하고 나머진 실패함
    """
    if sum(food_times) < k:
        return -1

    food_times = sorted([(value, index) for index, value in enumerate(food_times)], reverse=True) # food_times를 

    while True: # 서비스 장애가 발생하기 전까지 반복
        # 가장 적은 음식량 min_food가 0이 될 때까지 완주를 반복했을 때 걸리는 시간 = 최소 음식량 x 음식 개수 second가 소요됨
        if not food_times: # 중간에 food_times가 소진되면 -1반환 후 종료
            return -1

        runtime = (len_food := len(food_times)) * (min_food := food_times[-1][0]) 
        
        if k >= runtime: # k가 runtime보다 크면 runtime만큼 
            k -= runtime # runtime 만큼 시간을 빼주고
            food_times = [[i[0] - min_food, i[1]] for i in food_times if i[0] != min_food] # 이 경우 min_food만큼 남았던 음식은 다 먹어서 0이 될 터이니 전체 리스트에서 뺀다
            print(food_times, k)

        elif k < runtime:
            if k >= len(food_times): # k가 len_food보다 길면 회전을 더 할 수 있단 뜻이므로 k // len_food 번 더 회전하고 remainder를 가지고 다음 단계로 진행
                runs = k // len(food_times)
                k %= len(food_times)
                food_times = [[i[0] - runs, i[1]] for i in food_times]
                break

            elif k < len_food: # k가 len_food보다 작으면 회전을 더 할 수 없으므로 그대로 다음 단계로 진행. 이러면 최악의 경우 len(food)-1 번만큼 iterate할 것. 
                break
    # while문이 끝나면 현재 남은 k초 만큼 더 전진하고, 장애가 발생하면 장애가 발생한 시점 이후에서 가장 가까운 음식의 인덱스를 반환하는데, 
    # 이는 실질적으로 k + 1 초 만큼 전진해서 멈춘 그 자리의 인덱스를 반환하라는 것과 동치함. 
    print(f"remaining time={k}, {sorted(food_times, key=lambda x: x[1])}")

    food_times = deque([i for i in sorted(food_times, key=lambda x: x[1])])
    pos = 0
    while k >= 0:
        if not food_times:
            return - 1
        current_dish = food_times.popleft()
        k -= 1
        if current_dish[0] - 1 > 0: 
            food_times.append([current_dish[0]-1, current_dish[1]])
    else:
        return current_dish[1] + 1


def solution4(food_times, k):
    from collections import deque
    """
    """
    if sum(food_times) < k:
        return -1

    food_times = sorted([(value, index) for index, value in enumerate(food_times)], reverse=True) 
    time = 0
    last_food = [0, 0]

    while True: 
        # 가장 적은 음식을 모두 소모하기까지 최소 몇 개의 완전 회전이 가능한지 계산
        
        try: 
            current_food = food_times.pop()
        
        except:
            return -1
            
        # print(f"current_food = {current_food}")
        if current_food[0] == last_food[0]:
            continue

        required_time = (current_food[0] - last_food[0]) * len(food_times) + 1
        # print(f"required time: {required_time}, current time: {time}")

        if required_time + time <= k: 
            time += required_time
            last_food = current_food
            # print(f"full rotations made, time = {time}")

        else:
            food_times.append(current_food)
            if num_rotations := k // len(food_times) >= 1:
                food_times = [[i[0] - num_rotations, i[1]] for i in food_times]
                time += (num_rotations * len(food_times)) 
            break
    

    food_times = deque([i for i in sorted(food_times, key=lambda x: x[1])])
    pos = 0
    while time > 0:
        if not food_times:
            return -1
        current_dish = food_times.popleft()
        time -= 1
        if current_dish[0] - 1 > 0: 
            food_times.append([current_dish[0]-1, current_dish[1]])
    else:
        return current_dish[1] + 1
    

def solution5(food_times, k):
    """
    공식해답
    기존 솔루션들과 거의 똑같은 논리이되 heapq를 써서 우선순위 큐를 구현 후 만듦
    핵심은 요리들이 소진되는 시간을 초 단위로 시뮬레이션하는 것이 아니라 잔량이 가장 적은 요리 순서대로 소진된다는 점을 노려서 한 번에 요리들의 잔량 순서대로, 잔량 만큼 빼도록 하여서 
    전체 반복 횟수를 줄이는 데 있다. 

    이때 모든 음식을 우선순위 큐에 삽입한다. 이때 (음식 시간, 음식 번호) 형태의 튜플로 삽입한다. 가장 잔량이 적은 음식이 먼저 소진되므로, 우선순위 큐에 의해 자동으로 상향정렬된 요리들 중 
    가장 작은 것을 먼저 뺀다. 음식 1개가 온전히 소진되려면 (해당 음식의 잔량 x 전체 음식의 개수)초가 소진되므로 해당 시간을 계산해 총 시간은 K초에서 뺀다. 
    이런 방식으로 매 시행마다 음식들을 소진시켜가면서 네트워크 장애가 발생하기 직전까지 반복한다.  
    
    어떤 음식을 소진했을 때, 나머지 음식들의 잔량도 다 먹은 음식의 잔량만큼 줄어야 하나, 이미 heapq에 집어넣은 이상 값을 modify하기에는 비용이 크므로, 대
    previous라는 변수를 따로 만들어 이전 시행에서 먹은 잔량을 보관하도록 한다. now에는 현재 음식의 잔량을 저장하고, 따라서 특정 시행에서 먹는데 걸리는 시간은 (now - previous) * length
    가 된다. (length는 음식 개수)
    
    """
    import heapq 

    if sum(food_times) <= k: # 만약 모든 음식 잔량의 합이 K초보다 적다면 예외적으로 -1을 반환한다.
        return -1

    q = []
    for i in range(len(food_times)):
        heapq.heappush(q, (food_times[i], i + 1)) # 모든 음식을 (잔량, 번호) 튜플 형태로 우선순위 큐에 삽입한다. 

    sum_value = 0 # 먹기 위해 사용한 시간
    previous = 0 # 직전에 다 먹은 음식 시간
    length = len(food_times) # 남은 음식의 개수

    # ((q[0][0] - previous) * length)는 현재 잔량이 가장 적은 요리를 다 먹기 위해 소요되는 시간이므로, while 조건에 지금까지 사용한 시간 + 다음 요리를 다 먹을 시간을 미리 더해서 계산해본 뒤
    # 네트워크 장애 발생 이전에 다 먹을 수 있는 경우에만 먹는 걸로 처리함

    while sum_value + ((q[0][0] - previous) * length) <= k:
        now = heapq.heappop(q)[0] # 현재 요리 잔량
        sum_value += (now - previous) * length  
        length -= 1 # 한 요리를 온전히 소진했으면 요리 개수도 -1 
        previous = now # 이전 요리 잔량 갱신
    
    result = sorted(q, key = lambda x: x[1]) # 번호 기준으로 음식들을 다시 정렬
    return result[(k - sum_value) % length][1] # k초에 도달했을 때 몇 번째 음식 차례에 있는지 계산 후 반환
```
